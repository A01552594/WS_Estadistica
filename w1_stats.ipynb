# Descriptive statistics (History data)

*   Arithmetic mean (Media): Resumir rowdata con algo que representa a todas las observaciones 
*   Median (Mediana): La mediana ofrece una mejor representación del valor "medio" de una variable en comparación con la media aritmética cuando la distribución de los valores NO sigue una distribución normal
*   Mode (moda): El valor que más aparece

¿Usar mediana o moda?
En resumen, para una distribución sesgada, la mediana también será una mejor medida de la tendencia central. En realidad, en el caso de las variables con distribución normal, la mediana estará muy cerca de la media, por lo que la mediana puede ser siempre una buena medida de la tendencia central.


**Varianza y desviación estandar**
Ambos valores son usados para saber en promedio que tanto estan alejados los valores de la media

Varianza es normal y desviación estandar es la raiz cuadrada de la varianza
